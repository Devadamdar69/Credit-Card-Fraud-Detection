## Plotting the Correlation Matrix

```python
corrmat = df1.corr()
fig= plt.figure(figsize=(12,8))
sns.heatmap(corrmat,vmax=0.8, square=True)
plt.show()
```

```python
df['Normalized_amount']= scaler.fit_transform(df["Amount"].values.reshape(-1,1))
```

```python
df.drop(["Amount","Time"],inplace=True, axis=1)
df.head()
```

```python
#for all independent variable
x=df.drop(["Class"],axis=1)

y=df["Class"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
rf1 = RandomForestClassifier(n_estimators = 10)
```

```python
rf1.fit(xtrain, ytrain)
```

```python
prediction_rf = rf1.predict(xtest)
```

```python
rf_score = rf1.score(xtest,ytest)*100
rf_score
```

```python
outliers=len(fraud)
error=(prediction_rf != ytest).sum()
```

```python
accuracy = accuracy_score(ytest, prediction_rf)
accuracy
```

```python
precision = precision_score(ytest, prediction_rf)
precision
```

```python
recalling = recall_score(ytest, prediction_rf)
recalling
```

```python
f1= f1_score(ytest, prediction_rf)
f1
```

```python
MCC = matthews_corrcoef(ytest, prediction_rf)
MCC
```

# Visualizing the Confusion Matrix

```python
LABELS = ['Normal', 'Fraud']
conf_mat= confusion_matrix(ytest, prediction_rf)
plt.figure(figsize=(10,10))

sns.heatmap(conf_mat, xticklabels= LABELS, yticklabels= LABELS, annot=True, fmt='d')

plt.title("Confusion Matrix")
plt.ylabel("True Class")
plt.xlabel("Predicted Class")
plt.show()
```
